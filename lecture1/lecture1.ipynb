# Reminder: Kolmogorov and the probability calculus
In the 17th century Fermat and Pascal specifically invented the probability calculus to deal with the problem of physical uncertainty introduced by gambling. But it was then realized that that the concept of probability could be used in dealing also with the uncertainties introduced by ignorance. This lead in particular Bishop Butler to declare in the 18th century that "probability is the very guide to life". In this chapter we will give an introduction to the concept of probability and reasoning under uncertainty.

## Kolmogorov Axioms

In his book "Grundbegriffe der Wahrscheinlichkeitsrechnung" Kolmogorov describes the foundations of modern probability theory <cite data-cite="kolmogorov1933">(Kolmogorov, 1933)</cite> via set theory. Here we will introduce this formal language of probability in a very simple way using Venn diagrams.

Let $U$ be the universe of all possible events or outcomes of an experiment.
That is, if we are uncertain about which of a number of possibilities is true, we shall let $U$ represent all of them collectively. Then by definition the true event must lie within $U$ and the probability of finding the true event amongst all the events of $U$ must be maximal. By convention this maximum probability is set to $1$, giving us the first axium for probability theory <cite data-cite="kolmogorov1933">(Kolmogorov, 1933)</cite>: 

**Axiom 1.1** $P(U)=1$

This unit probability mass is distributed over $U$, perhaps evenly but perhaps also unevenly. For simplicity we assume that this probability mass is distributed evenly, meaning that the probability of any region is strictly proportional to its area. 

For any such region $X$ its area cannot be negative, even if $X$ would be empty. This yields the second of Komogorov's axioms:

**Axiom 1.2** *For all* $ X \subseteq U , P(X)\geq 0$

Note that this immediately implies:

**Theorem 1.1** *For all* $ X \subseteq U , P(X)\leq 1$.

Given these basic definitions we are interested in computing the probability of combined events $X$ and $Y$. If these events are mutually excluding then the result is trivial, giving us the third and last axiom of probability:

**Axiom 1.3** *For all* $ X,Y \subseteq U,\mathrm{if} \, X \cap Y = \emptyset,\mathrm{then} \, P(X\cup Y)= P(X) + P(Y)$

Therefore any function over a field of subsets of $U$ satisfying these three axioms will be a probability function.

The last axiom can be generalized towards adding the probability of events that overlap, i.e if $X$ and $Y$ intersect.
This yields the following theorem:

**Theorem 1.2** *For all* $ X,Y \subseteq U, P(X\cup Y)= P(X) + P(Y) - P(X\cap Y)$
 
 This can be understood intuitively from Figure 1.2. The area of $X\cup Y$ is less than the area of $X$ plus the area of $Y$ because when adding the area of intersection $X\cap Y$ has been counted twice. Therefore one needs to remove this excess area to find the the probability $P(X\cup Y)$ for any two events $X$ and $Y$.

### Conditional probabilities

  <img src="venn_diagrams_cond.png"/>
  <figcaption>Fig 1.2: (a) $P(X \cap Y)$; (b) conditional probability $P(X | Y)$ </figcaption>

The concept of conditional probability is crucial for the useful application of the probability calculus. We think of conditional probability as the probability of event $X$ being true given that event $Y$ is true. Note that this is distinct from the joint probability $P(X\cup Y)$, which is the probability that both, $X$ and $Y$ are true without knowing that one of them must be true. More specifically the conditional probability is defined as:

**Definition 1.1** $P(X|Y)=  \frac{P(X\cap Y)}{P(Y)}$

That is, given that the event $Y$ occured, or will occur, the probability that $X$ will also occur is $P(X|Y)$. This probability is clearly undefined if $Y$ is an event with zero probability. This is generally not a problem for **positiv** probability distributions, since, by definition they are non-zero over every event. 

A simple way to think about probabilities conditional upon $Y$ is to imagine that the universe of events U has shrunk to Y. The conditional probability of $X$ on $Y$ is then just the measure of what is left of $X$ relative to what is left of $Y$. In Fig. 1.2 this is indicated by the cross section representing $X\cap Y$ to the area of $Y$. This way of understanding conditional probability is justified by the fact that the conditional function $P(\cdot|Y)$ is itself a probability distribution.


### Independence $\def\ci{\perp\!\!\!\perp}$ 
Another important probability concept is that of independence (or marginal independence). Two events $X$ and $Y$ are probabilistically independent ( in notation $ X \ci{} Y$) whenever conditioning on one another leaves the probability of the uther unchanged.

**Definition 1.2** $ X \ci{} Y \equiv  P(X|Y) = P(X)$

This is provably symmetrical: $ X \ci{} Y \equiv Y \ci{} X$

### Conditional Independence 
Conditional independence generalizes the concept of independence to $X$ and $Y$ being independent given some additional event $Z$.

**Definition 1.3** $ X \ci{} Y|Z \equiv  P(X|Y,Z) = P(X|Z)$

This is a true generalization because $Z$ can ouf course be the empty set $\emptyset$. Conditional independence is a statement of redundant information. In particular conditional independence holds when the event $Z$ provides us with all the information on $X$ that $Y$ would too. Once you know $Z$ learning $Y$ would not provide you with new information. 

### Conditional probability theorems


**Theorem 1.3 Total Probability** Assume the set of events {A_i} is a partition of $U$; i.e. $\bigcup_i A_i = U$ and for any distinct $i$ and $j$ $ A_i\cap A_j = \emptyset $ Then: $ P(U) = \sum_i P(A_i) $

We can equally well partition the probability of any particular event $B$ instead of the whole event space $U$. In other words, under the above conditions (and if $\forall i; A_i \neq \emptyset$ ):
$ P(B) = \sum_i P(B\cap A_i) $

We shall refer to either formulation under the title "Total probability"


### The chain Rule

**Theorem 1.3 chain rule** Given three events $A$,$B$ and $C$ in a chain of influence 
$P(A,B,C)= P(A|B,C)\, P(B,C) = P(A|B,C)\, P(B|C) \, P(C) $ 

This theorem can be proofen by sequentially applying the definition of conditional probabilities.
The chain rule therefor trivially generalizes to

$P(\bigcap_{i=1}^{n} A_i) =  \prod_{i=1}^{n} P(A_i| \bigcap_{k=1}^{i-1} A_k)$

## What do we mean by plausibility?

Assume the following statement is true:

<center>$ A \,\,\,\,\,\,\, \implies \,\,\,\,\,\,\, B \,\,\,\,\,\,\,\,  $</center>
<center>      E.g. "It rained." $\implies$ "The street is wet."</center>


Then the following statement:

<center>$ \,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, B \,\,\,\,\,\,\, \implies \,\,\,\,\,\,\, A   $</center>
<center>      E.g. "The street is wet." $\implies$ "It rained."</center>


is neither true (T) nor false (F). In fact the proposition $A$, "It rained.", is one among many plausible explanations
for $B$,"The street is wet.".

However, the propositional calculus does not provide us with a way to quantify such a degree of belief.
In the next section we seek to extent the propositional calculus to deal with uncertainties and quantify plausibilities.

# Cox's Theorem: The plausibility calculus

Cox's theorem, derives the laws of probability theory, as stated by Kolmogorov's axioms above, from a set of reasonable postulates. This theorem justifies the interpretation of probability as degrees of plausibility.
Here we follow the proof of [van Horn 2003](https://www.sciencedirect.com/science/article/pii/S0888613X03000513)


<p>** Theorem **:</p>
<blockquote>
<p>Any system of logical reasoning that fulfills a certain set of desired conditions, as stated below, is isomorphic to probability theory.</p>
</blockquote>

<p>** Def 3 **:</p>
<blockquote>
<p>A state of information $X$ summarizes the information about some set of atomic propositions $A$, the basis of $X$, and their relationships to each other. $X$ is a logical closure of $A$, that is the union of $A$ and all compound propositions that involve only atomic propositions.</p>
</blockquote>

<p>** Def 4 **:</p>
<blockquote>
<p> We say that $A$ is equivalent to $B$ if ($A \Leftrightarrow B$) is a tautology (always true).</p>
</blockquote>

<p>** Def 5 **:</p>
<blockquote>
<p> If $X$ is a state of information and $A$ is a proposition in $X$ then:
$[A|X]$ is the plausibility of $A$ given $X$. We call $Y= A,X$ the joint state of information by adding the information that $A$ is true.</p>
</blockquote>

## Representation of plausibility

We want to measure degrees of plausibility by real numbers.


<p>** R 1 **:</p>
<blockquote>
<p> $[A|X]$, the plausibility $A$ given $X$, is a single real number. There exists a real number $T$ such that
    $[A|X]\leq T$ for all $A$ and $X$.</p>
</blockquote>


<p>** Def 6 **:</p>
<blockquote>
<p> We say: $X$ is consistent if there is no $A$ for which $[A|X]=T$ and $[\overline{A}|X]=T$.</p>
</blockquote>


## Compatibility with propositional calculus

We want the plausibility calculus to be consistent with logical reasoning.

<p>** R 2 **: Plausibility assignments are compatible with propositional calculus</p>
<blockquote>
<p> 1) If $A \Leftrightarrow X$ then $[A|X]=[B|X]$.
</blockquote>
<blockquote>
<p> 2) If $A$ is a tautology (always true) then $[A|X]=T$.
</blockquote>
<blockquote>
<p> 3) $[A|B,C,X]=[A|BC,X]$
</blockquote>
<blockquote>
<p> 4) If $X$ is consistent and $[\overline{A}|X]<T$ then $Y= A,X$ is consistent. 
</blockquote>

## Negation

<p>** R 3 **:</p>
<blockquote>
<p>  There exists a nonincreasing function $S_0$ such that:
    <center> $[\overline{A}|X] = S_0([A|X])$. </center>
</blockquote>

<p>** Proposition 1 **:</p>
<blockquote>
<p>  If $X$ is consistent and $x=[A|X]$ then:
    <center> $x = S_0(S_0(x))$. </center>
</blockquote>


<p>** Proof **:</p>
<blockquote>
<p> <center> $x = [\overline{\overline{A}}|X]=S_0([\overline{A}|X])=S_0(S_0([A|X])) \, , $. </center>
</blockquote>


<p>** Proposition 2 **:</p>
<blockquote>
<p>  Define $F=S_0(T)$. Then $F\leq [A|X] \leq T$ for all $A$ and consistent $X$.
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> <center> $[A|X]=S_0([\overline{A}|X])\geq S_0(T) \, , $. </center>
</blockquote>


## Universality

<p>** R 4 **:</p>
<blockquote>
<p>  There exists a nonempty set of real numbers $P_0$ with the following properties: 
    <blockquote>
    <p>1) $P_0$ is a dense subset of (F,T). That is for every pair of real numbers $a,b$ such that $F \leq a < b \leq T$, there exists some $c \in P_0$ auch that $a < c < b $ <br/><br/>
    <p>2)  For every $y_1,y_2,y_3 \in P_0$ there exists some consistent $X$ with a basis of at least three atomic propositione $A_1,A_2,A_3$ such that $[A_1|X]=y_1\, ,[A_2|A_1,X]=y_2\, ,[A_3|A_1,A_2,X]=y_3$.
    </blockquote>
</blockquote>   

<p>** Proposition 3 **:</p>
<blockquote>
<p>  $F<T$.
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> <center> $P_0$ is nonempty. </center>
</blockquote>


## Conjunction

<p>** R 5 **:</p>
<blockquote>
<p>  There exists a continuous function $H: [F,T]^2 \rightarrow [F,T]$ strictly increasing in both arguments on $(F,T]^2$ such that:<br/><br/>
     <center> $[AB|X] = H([A|B,X],[B|X])$ for any $A,B$ and consistent $X$  </center>
</blockquote>


## The product rule

<p>** Lemma 1 **:</p>
<blockquote>
<p>  $H(x,H(y,z)) = H(H(x,y),z)$
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> Let us consider $[ABC|X]$, where $X$ is consistent then use R 5, R 2:<br/><br/>
    <center> $\begin{eqnarray}[A(BC)|X] &=& H([A|(BC),X],[(BC)|X]) \nonumber \\ & =& H([A|B,C,X],H([B|C,X],[C|X]))\end{eqnarray}$  </center><br/><br/>
 alternative grouping yields <br/><br/>
    <center> $\begin{eqnarray}[(AB)C|X] &=& H([(AB)|C,X],[C|X]) \nonumber \\ & =& H(H([A|B,C,X],[B|C,X]),[C|X]))\end{eqnarray}$  </center><br/><br/>
but $A(BC) = (AB)C$ because of associativity. Now use  R 4:<br/><br/>
    <center>$H(x,H(y,z))=H(H(x,y),z)$ for all $x,y,z \in P_0$</center><br/><br/>
Since $H$ is continous and $P_0$ is dense this holds for all $x,y,z \in [F,T]$
</blockquote>


<p>** Lemma 2 **:</p>
<blockquote>
<p>  Let $a$ and $b$, with $a<b$, be real numbers. Suppose that $f:(a,b]^2 \rightarrow (a,b]  $ is a continous function strictly increasing in both arguments, and satisfies the associative equation:<br/><br/>
    <center>$f(x,f(y,z))=f(f(x,y),z)$ for all $x,y,z \in (a,b]$</center><br/><br/>
    Then there is a strictly increasing, continous function $g$ such that:<br/><br/>
    <center>$g(f(x,y))= g(x) + g(y)$ for all $x,y,z \in (a,b]$</center>
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> see J. Aczel, Lectures on Functional Equations and their Application, Academic press 1996, p. 256
</blockquote>

Now we are ready to get the product rule

<p>** Lemma 3 **:</p>
<blockquote>
<p> There exists a continuous, strictly increasing nonnegative function w, such that:<br/><br/>
    <center>$w([AB|X])=w([A|B,X])\,w([B|X])$ for every $A,B$ and consistent $X$</center>
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> R 5 and Lemma 1 allow for Lemma 2 with $f = H$, $a=F$ and $b=T$.
    Define $w(x)=\mathrm{exp}(g(x))$ then:<br/> <br/>
    <center>$w(H(x,y))=w(x)\,w(y)$ for every $x,y \in (F,T]$</center> <br/>
    Since $w$ is increasing, continous and nonnegative on $(F,T]$, $\lim_{x \to F} w(x)$ exists and is nonnegative. Define $w(F)=\lim_{x \to F} w(x)$ (makes $w$ increasing, continuous and nonnegative on $[F,T]$). By continuity of $w$ at $F$ we then have: <br/><br/>
    <center>$w(H(x,y))=w(x)\,w(y)$ for every $x,y \in [F,T]$</center> <br/>
    and combination with R 5 gives us the Lemma. <br/>
    <br/>
</blockquote>

Are we getting the correct ranges for $w$?

<p>** Lemma 4 **:</p>
<blockquote>
<p> $w(F)=0$,$w(T)=1$ and $0<w(x)<1$ for $F<x<T$
</blockquote>

<p>** Proof **:</p>
<blockquote>
<p> For $x>F$ we have $w(x)>w(F)\geq 0$.
    Choose any $A$, tautology $D$, and consistent $X$ with $[A|X]>F$ then: <br/> <br/>
    <center>$\begin{eqnarray} w([A|X])&=&w([DA|X]) \nonumber \\ &=& w([D|A,X])w([A|X])\nonumber \\ &=& w(T)w([A|X]) \end{eqnarray}$. </center> <br/>
    Therefore $w(T)=1$ and hence $w(x)<1$.<br/> <br/> 
    Now let $w(F)=z>0$. Then $0<z<1$ hence $z<\sqrt(z)<1$.<br/>
    Choose $A_1,A_2$ consistent $X$, and $x \in P_0$ such that $w(z)^{-1}<x<w(\sqrt(z))^{-1}$ and $[A_1|X]=[A_2|A_1,X]=x$ then: <br/> <br/>
    <center>$w([A_1,A_2|X])=w([A_2|A_1,X])w([A_1|X]) = w(x)^2 < z = w(F)$. </center> <br/><br/>
    but that means $[A_1,A_2|X]<F$ a contradiction!!!<br/>
    Thus $w(F)=0$!
</blockquote>

## Cox's theorem

Let us rename $w(x)=p(x)$

<p>** Cox's theorem **: There exists a continuous strictly increasing function p, such that for every $A,B$ and consistent $X$ </p>
<p> 1) $p(A|X) = 0$ iff $[A|X]= F$
<p> 2) $p(A|X) = 1$ iff $[A|X]= T$
<p> 3) $0\leq p(A|X) \leq 1$
<p> 4) $p(AB|X) = p(A|X)\,p(B|A,X)$
<p> 5) $p(\overline{A}|X) = 1 - p(A|X)$ if $X$ is consistent.


<br/>
Therefore, any system of plausible inference that satisfies R 1 to R 5, and hence common sense reasoning, is isomorphic to probability theory.
<br/><br/>
We can therefore think of probabilities as degrees of plausibility or belief.
<br/><br/>
Breaking the laws of probability leads to irrational behavior, e.g. accepting a bet with a guaranteed loss (see Dutch Books).

## Bayes' Theorem
The origin of the Bayesian philosophy lies in an interpretation of Bayes' theorem. <br/><br/>
Her we discuss the derivation and application of the theorem.

# Bayes' law
Bayes's law can be directly derived from the product rule of the probability/plausibility calculus. <br/><br/>

<center> $P(A|B,X)\,P(B|X)=P(A,B|X)=P(B|A,X)\,P(A|X)$ </center>

From that we obtain Bayes's law by rearranging the left and right handsides of the above equation:<br/><br/>

<center> $P(A|B,X)=P(A|X)\,\frac{P(B|A,X)}{P(B|X)}$ </center>

We then call:
<blockquote>
<p> 1) $P(A|B,X)$ the posterior of $A$ given $B$ and some background information $X$.
</blockquote>
<blockquote>
<p> 2) $P(A|X)$ the prior of $A$ in light of some background information $X$.
</blockquote>
<blockquote>
<p> 3) $P(B|A,X)$ the likelihood of the observation $B$ given a proposition $A$ and some background information $X$
</blockquote>
<blockquote>
<p> 4) $P(B|X)$ the normalization or evidence
</blockquote>

<br/><br/>

This result is not controversial, it is just a "simple" theorem of plausibility.

However the following claim that conditionalization is a right and proper way to adjust degrees of plausibility is controversial:

<p>** Def 6 **: Conditionalization</p>
<blockquote>
<p> After applying Bayes' theorem to obtain $P(A|B,X)$ adopt that as the posterior degree of plausibility or belief of $A$.</p>
</blockquote>


More specifically conditionalization advocates belief updating via probabilities conditional upon available evidence. It identifies posterior probability with conditional probability. 
 
This may also not seem controversial, but there exist certain situations in which conditionalization does not work:

1) There exists the joint prior $P(A,B|X)$
2) There exists the joint prior $P(B|X)$ and $P(B|X)>0$

Without this conditionalization is a non-starter. <br/><br/>

Bayes' theorem receives major criticism due to the prior. The prior $P(A|X)$ assigns a degree of plausibility to a certain outcome of an experiment prior to the actual fact. <br/><br/>

"Where do these numbers come from?" Clark Glymour 1980 <br/><br/>

Answer: <br/><br/>

"It makes no difference what sources your priors have. Non-extreme priors converge in the limit." Bruno de Finetti 1964 <br/><br/>


We will discuss different approaches to design priors later in the lecture.






# Is the posterior a probability?

To check if the posterior is a probability we apply the probability theorems.

** Check sum rule: ** <br/><br/>

<center> $\begin{eqnarray} 1 &=& P(A|B,X) + P(\overline{A}|B,X) \nonumber \\ &=& \frac{P(A|X)P(B|A,X)}{P(B|X)} + \frac{P(\overline{A}|X)P(B|\overline{A},X)}{P(B|X)} \nonumber \\   \end{eqnarray}$ </center>

Therefore for existing $P(B|X)>0$ <br/><br/>

<center> $P(B|X)=P(A|X)P(B|A,X) + P(\overline{A}|X)P(B|\overline{A},X)$ </center><br/><br/>

the posterior solves the sum rule

** Check product rule: ** <br/><br/>
The proof is trivial. For existing $P(B|X)>0$ the posterior fulfills the product rule by design.


** Check $0\leq P(A|B,X) \leq 1$ : ** <br/><br/>
For existing $P(B|X)>0$ : <br/><br/>

<center> $ 0\leq P(A,B|X) \leq 1 $ </center><br/><br/>

Division by $P(B|X)$ yields:<br/><br/>

<center> $ \begin{eqnarray} 0 \leq &\frac{P(A,B|X)}{P(B|X)}& \leq \frac{1}{P(B|X)} \nonumber \\ 0\leq &P(A|B,X)& \leq \frac{1}{P(B|X)} \end{eqnarray}$ </center> <br/><br/>

Thus $0\leq P(A|B,X)$. Analogously we obtain $0\leq P(\overline{A}|B,X)$. Then: <br/><br/> 

<center> $ P(A,B|X) = 1 - P(\overline{A}|B,X)  \leq 1 $ </center><br/><br/>

Since the sum rule holds and $0\leq P(\overline{A}|B,X)$ the statement is proven. <br/><br/>


Therefore posterior is a probability if:

1) $P(A,B|X)$ exists and is a probability. <br/><br/>
2) $P(B|X)=P(A|X)P(B|A,X) + P(\overline{A}|X)P(B|\overline{A},X)>0$. <br/><br/>


## Bayesian learning

Assume we have observations $B_1,B_2$ then frequen application of Bayes's law yields:

<center> $ \begin{eqnarray} P(A|B_1,B_2,X) &=& \frac{ P(A|X)\,P(B_1,B_2|A,X)}{P(B_1,B_2|X)} \nonumber \\ &=&\frac{ P(A|X)\,P(B_1|A,X)}{P(B_1|X)} \frac{\,P(B_2|B_1,A,X)}{P(B_2|B_1,X)} \nonumber \\ &=& P(A|B_1,X) \frac{\,P(B_2|B_1,A,X)}{P(B_2|B_1,X)}\end{eqnarray}$ </center> <br/><br/>

Where in the last line we see, that the new posterior $P(A|B_1,B_2,X)$ is just the previous posterior $P(A|B_1,X)$ multiplied with the likelihood of the new observation $B_2$. Bayesian learning is therefore the sequential updating of the posterior distribution in light of new observations. <br/><br/>

A generalization to arbitrary many observations $B_N$ is left as a homework.


## Extending Bayes rule to many propositions

Say $A_1,A_2, ..., A_N$ is an exhaustive set of all possible explanations of $B$. Then the respective priors $P(A_i|X)$ fulfil the sum rule:

<center> $ \sum_i P(A_i|X) = 1$ </center> <br/><br/>

Then the respective posteriors  $P(A_i|B,X)$ are given as:

<center> $P(A_i|B,X) = \frac{P(A_i|X) P(B|A_i,X)}{\sum_j P(A_j|X)P(B|A_j,X)}$ </center> <br/><br/>

## Bayesian Hypothesis testing

Philosophy of science: <br/><br/>

<center> "We do not rule out models, we just determine their relative plausibilities." </center> <br/><br/>

Note that this is a direct instruction to design hypotheses tests in a Bayesian scenario. The posterior quantifies the degree of plausibility or belief that a proposition $A_1$ is true. We may now wonder if proposition $A_1$ is more plausible than a second propositione $A_2$. To do this we just estimate the ratios of the two posteriors:

<center> $\begin{eqnarray}\frac{P(A_1|B,X)}{P(A_2|B,X)} = \frac{P(A_1|X)\,P(B|A_1,X)}{P(A_2|X)\,P(B|A_2,X)}\frac{P(B|x)}{P(B|x)}\end{eqnarray}$ </center> <br/><br/>

We call:

<center> $O(A_1,A_2|X)=\frac{P(A_1|X)}{P(A_2|X)} $<br/><br/>

the prior odds and:

<center> $O(A_1,A_2|B,X)=\frac{P(A_1|B,X)}{P(A_2|B,X)} $<br/><br/>
    
the posterior odds.

Then we can write the ratio of the posterior and the prior odds as:

<center> $\frac{O(A_1,A_2|B,X)}{O(A_1,A_2|X)}=\frac{P(B|A_1,X)}{P(B|A_2,X)} = BF_{1,2} $<br/><br/>
    
Where we call $BF_{1,2}$ the Bayes factor. It is clear that Bayes' factor in favour of $A_1$ will only rise if and only if the likelihood ratio is greater than one.

To interpret these values Jeffreys introduced the following scale scale


| $BF$   | Strength of evidence    | 
| ----------------------- |:-----------------------:| 
| $10^0$ to $10^{1/2}$    | barely worth mentioning |
| $10^{1/2}$ to $10^{1}$  | substantial             |
| $10^1$ to $10^{3/2}$    | strong                  |
| $10^{3/2}$ to $10^{2}$  | very strong             |
| $>10^{2}$               | decisive                |

   
    
# Some Advantages of Bayesian Hypothesis testing

Bayes factors and posterior model probabilities describe the relative support or plausibility for a set of candidate hypotheses. Other advantages of Bayesian hypothesis testing include the following:

** 1) guaranteed coherence **

Suppose we have a set of three candidate hypotheses, $A_1$,$A_2$ and $A_3$. Then:

$\begin{equation}
\frac{P(A_1|B,X)}{P(A_3|B,X)} = \frac{P(A_1|B,X)}{P(A_2|B,X)}\, \frac{P(A_2|B,X)}{P(A_3|B,X)}  
\end{equation}
$

In terms of Bayes factors this reads:

$\begin{equation}
BF_{1,3} = BF_{1,2}\, BF_{2,3} 
\end{equation}
$

For instance, when the data are five times as likely to occur under $A_1$ than under $A_2$, and seven time as likely under $A_2$ than under $A_3$, it follows that the data are $5 \times 7 = 35$ times as likely under $A_1$ than under $A_3$. 

This is a genuine Bayesian result and no comparable result exists in classical statistics.

** 2) Evidence can be obtained in favour of the Null hypothesis **
Bayesian hypothesis testing allows one to obtain evidence in favor of the null hypothesis. Because theories and models often predict the absence of a difference, it is vital for scientific progress to be able to quantify evidence in favor of the null hypothesis.


## Example: Bayesian reasoning

Example taken from:

* Kevin B. Korb, Ann E. Nicholson, *Bayesian Artificial Intelligence, Second Edition*, Chapman & Hall/CRC, 2010 <br/><br/>

Suppose the women attending a particular clinic show a long-term chance of 1 in 100 of having breast cancer. Suppose also that the initial screening test used at the clinic has a false positive rate of 0.2 (that is, 20 percent of women without cancer will test positive for cancer) and that it has a false negative rate of 0.1 ( that is 10 percent of women with cancer will test negative). The laws of probability dictate from these facts that the probability of a positive test given cancer is 90 percent. Suppose you have been tested positive, what is the probability of having cancer?

** Probability of a positive test ** <br/><br/>
So clearly the probability of testing positive given that a patient has cancer is:
$\begin{eqnarray}
P(Pos| cancer) &=& 0.9 \nonumber \\
\end{eqnarray}
$

** What is the chance of having cancer **<br/><br/>
But this does not answer the question most of us would like to have answered and that is what is the probability of having cancer given a positive test. Note that this is not the same question. In analogy to propositional calculus
a positive test does not equal to a patient having cancer.  

$\begin{eqnarray}
P(Cancer| Pos) &=& \frac{P(Pos | Cancer) P(Cancer)}{P(Pos)} \nonumber \\
&=& \frac{P(Pos | Cancer) P(Cancer)}{P(Pos|Cancer) P(Cancer) + P(Pos|\overline{Cancer}) P(\overline{Cancer})} \nonumber \\
&=& \frac{0.9 \times 0.01}{0.9 \times 0.01 + 0.2 \times 0.99}\nonumber \\
&\approx& 0.043 \nonumber \\
\end{eqnarray}
$

Now the discrepancy between 4 percent and 80 or 90 percent is no small matter particularly if the consequences of an error involves unnecessary surgery or (in the reverse case) leaving a cancer untreated.
